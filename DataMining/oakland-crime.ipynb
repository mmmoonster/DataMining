### 检查重复数据信息

```python
def dupdata(rec):
    lenr = len(rec)
    typer = rec.duplicated().any()
    print(lenr, typer)
    return
```

```python
dupdata(rec1)
dupdata(rec2)
dupdata(rec3)
dupdata(rec4)
dupdata(rec5)
dupdata(rec6)
```

## 标称属性分析及可视化展示

```python
def stans(num, datalen):
    print("number of classes:", len(num))
    print("number of missing data:", datalen-num.sum())
    print("the top 10 classes:\n", num[0:10])
    num[0:10].plot(kind="bar", figsize=(20,9))
    return
```
